```python
from openai import OpenAI
import os
```

```python
API_key = os.getenv('OPENAI_API_KEY')
if not API_key: raise ValueError("‚ùå Key has no value!")

client = OpenAI(api_key=API_key)
```

```python
response = client.responses.create(
  model="gpt-4.1-nano",
  input="say hello!",
  store=False,
)
```

```python
print(response.output_text);
```
